```python
import sys
import collections

# Helper functions
def toBase(num,base):
    base_num = ""
    while num > 0:
        dig = int(num%base)
        if dig < 10:
            base_num += str(dig)
        else:
            base_num += chr(ord('A')+dig-10)
        num //= base
    base_num = base_num[::-1]
    return base_num
  
def triToEye(num):
  return toBase(num, 5).rjust(3, "0")

def eyeToTri(eye):
  return int(eye, 5)

def rotEye(e):
  return e[1] + e[2] + e[0]

def eyeToTri_unrotated(e):
  t1 = eyeToTri(e)
  t2 = eyeToTri(rotEye(e))
  t3 = eyeToTri(rotEye(rotEye(e)))
  return min(min(t1, t2), t3)

def eyeToTri_flip(e):
  return eyeToTri(e[0] + e[2] + e[1])
```

```python
# Unique ID helper
class UUID:
  def __init__(self):
    self.next = 0
    self.map = { }
    self.invMap = { }

  def getUniqueID(self, val):
      if val not in self.map:
        self.map[val] = self.next
        self.invMap[self.next] = val
        self.next += 1
      return self.map[val]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter, OrderedDict
from statistics import mean, median
from matplotlib import pyplot as plt
from IPython.display import Javascript
```

```python
def printAscii(messages, offset=32):
  # Output messages as ascii with v + 32
  print("Ascii of message (1 - " + str(len(messages)) + ")")
  print("-------------------------")
  print("Note: This will break for messages with values > 95\n")
  for msg in messages:
    print("".join([ chr(v + offset) for v in msg ]))


def showMatching(messages, aspect=False):
  im = np.zeros((len(messages), max([ len(msg) for msg in messages ])), 'i4')
  for y, msg in enumerate(messages):
    for x, l in enumerate(msg):
      im[y, x] = l

  print("Aligned sections of messages")
  print("----------------------------\n")
  ax = plt.figure(0, (24, 6)).subplots()
  ax.set(xlabel="Message Index", ylabel="Message")
  if aspect:
    ax.imshow(im, interpolation='nearest', aspect='auto')
  else:
    ax.imshow(im, interpolation='nearest')
  plt.show()
```

```python
def getIoC(msg, letterCount):
  # Calculate IoC of the message
    N = len(msg)
    freqs = collections.Counter(msg)
    alphabet = range(letterCount)
    freqsum = 0.0
    for l in alphabet:
        freqsum += freqs[l] * (freqs[l] - 1)
    return freqsum / ((N * (N - 1)) / letterCount)


def fullIoC(messages, letterCount=None):
  # Calculate letter count
  if letterCount == None:
    letters = set()
    for msg in messages:
      for eye in msg:
        letters.add(eye)
    letterCount = len(letters)

  # Output overview of IoC of each + all
  print("IoC Oveview")
  print("-----------\n")
  print("Letters (" + str(letterCount) + "): " + str(letters))

  full = []
  print("\nIOC of each:")
  for i, msg in enumerate(messages):
    print(str(i + 1) + ": " + str(getIoC(msg, letterCount)))
    full += msg
  print("all: " + str(getIoC(full, letterCount)))
```

```python
def letterFrequency(messages):
  # ---------------------------
  # PULLED FROM ANALYSIS COLAB
  # ---------------------------
  # - Letter frequency overall
  # ---------------------------

  counts = Counter()
  for msg in messages:
    counts += Counter(msg)
  countsSorted = counts.most_common()

  print("Letter Frequencies (Overall)")
  print("----------------------------\n")
  print(f"5 most common letters: { countsSorted[:5] }")
  print(f"5 least common letters: { countsSorted[-5:] }")
  print(f"Median frequency: {median(counts.values())}")
  print(f"Mean frequency: {mean(counts.values())}")
  print("")

  x = range(max(counts) * 2)
  y = [counts[a] for a in x]

  ax1, ax2 = plt.figure(0, (18, 4)).subplots(1, 2)
  ax1.bar(x, y)
  ax1.set(xlabel="Cipher Letter", ylabel="Count")
  ax2.bar(x, sorted(y, reverse=True))
  ax2.set(xlabel="Cipher Letter", ylabel="Count")
  plt.show()


def letterFrequencyMessage(messages):
  # ---------------------------
  # PULLED FROM ANALYSIS COLAB
  # ---------------------------
  # - Letter frequency per message
  # ---------------------------

  m = max([ max(msg) for msg in messages ]) + 1
  im = np.zeros((len(messages), m), 'i4')
  for y, msg in enumerate(messages):
    for let in msg:
      im[y, let] += 1

  print("Letter Frequencies (Per message)")
  print("--------------------------------\n")
  ax = plt.figure(0, (16, 4)).subplots()
  ax.set(xlabel="Cipher Letter", ylabel="Message")
  ax.imshow(im)
  plt.show()
```

```python
def kappa_test(msg1, msg2, offset):
    """
    Kappa test for rate of coincidence between distinct ciphertexts.
    Test the two ciphertexts and return an (N, D) tuple, where:
        N is the number of coincidences
        D is the number of tests performed
    """
    m1 = msg1[offset:]
    m2 = msg2[:len(m1)]
    m1 = m1[:len(m2)]
    return sum(a == b for a,b in zip(m1, m2)), len(m1)


def kappaPeriodic(messages):
  # ---------------------------
  # PULLED FROM ANALYSIS COLAB
  # ---------------------------
  # - Kappa test (Periodic)
  # ---------------------------

  bounds = (4, 90)
  x = list(range(*bounds))
  results_y = []
  for offset in x:
      matches = 0
      checks = 0
      for msg1 in messages:
          for msg2 in messages:
              mtch, check = kappa_test(msg1, msg2, offset)
              matches += mtch
              checks += check

      results_y.append(1000 * matches / checks)

  print("Kappa Periodic Test")
  print("-------------------\n")
  ax = plt.figure(0, (12, 4)).subplots()
  ax.bar(x, results_y, 0.8, label="Coincidences per 1000")
  ax.plot(bounds, (66, 66), 'g', label="Expected (English)")
  ax.plot(bounds, (12, 12), 'r', label="Expected (Random)")
  ax.set(xlabel="Period Length", ylabel="Count")
  ax.legend()
  plt.show()


def kappaAuto(messages):
  # ---------------------------
  # PULLED FROM ANALYSIS COLAB
  # ---------------------------
  # - Kappa test (Autocorrelation)
  # - (unrotated unique)
  # ---------------------------

  bounds = (1, 40)
  x = list(range(*bounds))
  results_y = []
  for w in x:
      matches = 0
      checks = 0
      for i in range(len(messages)):
          match, check = kappa_test(messages[i], messages[i], w)
          matches += match
          checks += check

      results_y.append(1000 * matches / checks)

  print("Kappa Auto-Correlation Test")
  print("---------------------------\n")
  ax = plt.figure(0, (12, 4)).subplots()
  ax.bar(x, results_y, 0.8, label="Coincidences per 1000")
  ax.plot(bounds, (66, 66), 'g', label="Expected (English)")
  ax.plot(bounds, (12, 12), 'r', label="Expected (Random)")
  ax.set(xlabel="Offset", ylabel="Count")
  ax.legend()
  plt.show()
```


# Variant: Mirrored

---

```python
def mirrorVert(eye):
  if int(eye) == 1: return 3
  elif int(eye) == 3: return 1
  else: return eye


def mirrorHorz(eye):
  if int(eye) == 2: return 4
  elif int(eye) == 4: return 2
  else: return eye


def visualRotate(eye):
  if int(eye) == 1: return 3
  elif int(eye) == 2: return 4
  elif int(eye) == 3: return 1
  elif int(eye) == 4: return 2
  else: return eye


def mirrorEye(eyeFlat, variant):
  if variant == 1: return "".join([ str(mirrorVert(mirrorHorz(single))) for single in eyeFlat ])
  elif variant == 2: return "".join([ str(mirrorVert(single)) for single in eyeFlat ])
  elif variant == 3: return "".join([ str(mirrorHorz(single)) for single in eyeFlat ])
  elif variant == 4: return "".join([ str(visualRotate(single)) for single in eyeFlat ])
  elif variant == 5:
    mirrored = [ str(mirrorVert(single)) for single in eyeFlat ]
    return mirrored[1] + mirrored[0] + mirrored[2]
  elif variant == 6:
    mirrored = [ str(visualRotate(single)) for single in eyeFlat ]
    return mirrored[1] + mirrored[0] + mirrored[2]
  else: return None


eyeMessagesMirrored1Tri = [ [ eyeToTri( mirrorEye(eye, 1) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
eyeMessagesMirrored2Tri = [ [ eyeToTri( mirrorEye(eye, 2) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
eyeMessagesMirrored3Tri = [ [ eyeToTri( mirrorEye(eye, 3) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
eyeMessagesMirrored4Tri = [ [ eyeToTri( mirrorEye(eye, 4) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
eyeMessagesMirrored5Tri = [ [ eyeToTri( mirrorEye(eye, 5) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
eyeMessagesMirrored6Tri = [ [ eyeToTri( mirrorEye(eye, 6) if i % 2 == 1 else eye ) for i, eye in enumerate(msg) ] for msg in eyeMessagesFlat ]
```

```python
# Get diagonal letters
def diagonalEye(eye, alt):
  return [ int(eye[0]), int(eye[2]), int(eye[1]) ] if not alt else [ int(eye[1]), int(eye[2]), int(eye[0]) ]

def diagonalMessage(msg):
  newMsg = []
  for i, eye in enumerate(msg):
    newMsg += diagonalEye(eye, i % 2 == 1)
  return newMsg

eyeMessagesDiagonal = [ diagonalMessage(msg) for msg in eyeMessagesFlat ]
```

```python
from collections import Counter

print("\nDiagonal Count")
print("----------------\n")
counters = [ Counter(msg) for msg in eyeMessagesDiagonal ]
print(counters[0])
print(counters[1])
print(counters[0][0] + counters[0][1] + counters[0][3] + counters[0][4])
print(counters[1][0] + counters[1][2] + counters[1][3] + counters[1][4])
```

```python
# Get diagonal overlapped
eyeMessagesDiagonalOverlapped = []
eyeMessagesDiagonalOverlappedUnique = []

uuid = UUID()
for i in range(5):
  east = eyeMessagesDiagonal[i * 2]
  west = eyeMessagesDiagonal[i * 2 + 1] if i < 4 else []

  overlapped = []
  overlappedUnique = []
  for o in range(max(len(east), len(west))):
    value = ""
    if o < len(east):
      value = (str(east[o]) + value) if (value == "" or east[o] < int(value)) else (value + str(east[o]))
    if o < len(west):
      value = (str(west[o]) + value) if (value == "" or west[o] < int(value)) else (value + str(west[o]))
    if len(value) == 1:
      value = value + value

    # Value will like the following:
    #   0 on 0 = 00
    #   0 on 1 = 01
    #   1 on 0 = 01
    #   4 on 2 = 24
    #   2 on 2 = 22
    #   3 on None = 33

    overlapped.append(value)
    overlappedUnique.append(uuid.getUniqueID(value))

  eyeMessagesDiagonalOverlapped.append(overlapped)  
  eyeMessagesDiagonalOverlappedUnique.append(overlappedUnique)
```

```python
print("\nDiagonal Overlapped")
print("--------------------\n")
for msg in eyeMessagesDiagonalOverlapped:
  print(msg)

print("")
print(uuid.map)
print("".join([ chr(v + 65) for v in range(15) ]))

print("")
for msg in eyeMessagesDiagonalOverlappedUnique:
  print("".join([ chr(v + 65) for v in msg ]))
```
