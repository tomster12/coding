```python
from locallib import analysis, eyeutil
import matplotlib.pyplot as plt

```

```python
analysis.plot_im(
    im=analysis.conv_msgs_to_im(eyeutil.msgs_eye_tri),
    title="Plain Messages (CIPHERTEXT_NUM)",
    figsize=(28,2))
plt.show()

analysis.plot_im(
    im=analysis.conv_msgs_to_im(eyeutil.msgs_eye_tri),
    ascii=True,
    title="Plain Messages (CIPHERTEXT_ASCII)",
    figsize=(28,2))
plt.show()
```

```python
shared_im = analysis.calc_shared(msgs=eyeutil.msgs_eye_tri, zero_value=-1)
analysis.plot_im(
    im=shared_im,
    labels=eyeutil.msgs_eye_tri,
    title="Shared Sections",
    to_dull=True,
    under_value=0,
    figsize=(28,2))
plt.show()
```

```python
gap_im_start = analysis.calc_gaps(msgs=eyeutil.msgs_eye_tri, limit=16, zero_value=-1)
analysis.plot_im(
    im=gap_im_start,
    labels=eyeutil.msgs_eye_tri,
    title="Gaps (max 16, start only, msg value)",
    to_dull=True,
    figsize=(28,2))
plt.show()

gap_im_both = analysis.calc_gaps(msgs=eyeutil.msgs_eye_tri, limit=16, zero_value=-1, include_end=True)
analysis.plot_im(
    im=gap_im_both,
    labels=eyeutil.msgs_eye_tri,
    title="Gaps (max 16, start+end, msg value)",
    to_dull=True,
    figsize=(28,2))
plt.show()

analysis.plot_im(
    im=gap_im_both,
    ascii=True,
    labels=eyeutil.msgs_eye_tri,
    title="Gaps (max 16, start+end, msg value ASCII)",
    to_dull=True,
    figsize=(28,2))
plt.show()

gap_im_both_nan = analysis.calc_gaps(msgs=eyeutil.msgs_eye_tri, limit=16, include_end=True)
analysis.plot_im(
    im=gap_im_both,
    labels=gap_im_both_nan,
    title="Gaps (max 16, start + end, gap size)",
    to_dull=True,
    figsize=(28,2))
plt.show()
```

```python
im_repeats = analysis.calc_repeats(eyeutil.msgs_eye_tri)
analysis.plot_im(
    im=im_repeats,
    title="Eyes Repeats (REPEAT_COUNT)")
plt.show()

msgs_random, _ = eyeutil.generate_random_msgs_tri()
im_repeats = analysis.calc_repeats(msgs_random)
analysis.plot_im(
    im=im_repeats,
    title="Repeats Random (REPEAT_COUNT)")
plt.show()
```

```python
isos_19 = analysis.calc_isomorphs(eyeutil.msgs_eye_tri[0:9])
isos_13 = analysis.calc_isomorphs(eyeutil.msgs_eye_tri[0:3])
isos_46 = analysis.calc_isomorphs(eyeutil.msgs_eye_tri[3:6])
isos_79 = analysis.calc_isomorphs(eyeutil.msgs_eye_tri[6:9])
```

```python
isos_im_19 = analysis.conv_isomorphs_to_img(eyeutil.msgs_eye_tri[0:9], isos_19, 0)
isos_im_13 = analysis.conv_isomorphs_to_img(eyeutil.msgs_eye_tri[0:3], isos_13, 0)
isos_im_46 = analysis.conv_isomorphs_to_img(eyeutil.msgs_eye_tri[3:6], isos_46, 0)
isos_im_79 = analysis.conv_isomorphs_to_img(eyeutil.msgs_eye_tri[6:9], isos_79, 0)

analysis.plot_im(
    im=isos_im_19,
    to_dull=True,
    under_value=0.1,
    labels=eyeutil.msgs_eye_tri[0:9],
    title="(1-9) Isomorphs",
    figsize=(28,2))

analysis.plot_im(
    im=isos_im_13,
    to_dull=True,
    under_value=0.1,
    labels=eyeutil.msgs_eye_tri[0:3],
    title="(1-3) Isomorphs",
    figsize=(28,2))

analysis.plot_im(
    im=isos_im_46,
    to_dull=True,
    labels=eyeutil.msgs_eye_tri[3:6],
    title="(4-6) Isomorphs",
    figsize=(28,2))

analysis.plot_im(
    im=isos_im_79,
    to_dull=True,
    under_value=0.1,
    labels=eyeutil.msgs_eye_tri[6:9],
    title="(7-9) Isomorphs",
    figsize=(28,2))

plt.show()
```
