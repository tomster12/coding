```python
# e1: 4040
# w1: 4124
# e2: 4754
# w2: 4295
# e3: 5656
# w3: 4748
# e4: 5385
# w4: 4936
# e5: 4545

values = [4040,
          4124,
          4754,
          4295,
          5656,
          4748,
          5385,
          4936,
          4545]


def prime_factors(n):
    factors = {}
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1
    if n > 1:
        factors[n] = factors.get(n, 0) + 1
    return factors


for x in values:
    factors = prime_factors(x)
    expanded = []
    for p, count in sorted(factors.items()):
        expanded.extend([str(p)] * count)
    print(f"{x}: {'*'.join(expanded)}")
```
