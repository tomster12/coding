```python
from locallib import eyeutil
from matplotlib import pyplot as plt
from collections import Counter
```

```python
messages, _ = eyeutil.generate_random_msgs_tri(83)
# messages = eyeutil.msgs_eye_tri
```

```python
down_counter = Counter()
up_counter = Counter()

for message in messages:
    for i in range(0, len(message)):
        if i % 2 == 0:
            down_counter[message[i]] += 1
        else:
            up_counter[message[i]] += 1


difference = Counter()
for key in down_counter.keys():
    difference[key] = abs(down_counter[key] - up_counter[key])

# Plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(difference.keys(), difference.values())
ax.set_xlabel('Message')
ax.set_ylabel('Difference')
ax.set_title('abs(freq(down) - freq(up))')
plt.show()
```
